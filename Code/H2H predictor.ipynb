## Observation: We need to look at stats that we know before matches. The current set of features are static: halftime or post-match results

## 1. Static Model: to inform which stats might be more important, let's first take a look at logistic regression using the static features

```python
# import LogisticRegressionCV function from sci-kit learn

from sklearn.linear_model import LogisticRegression
```

```python
# Break data into features and target (outcome variable)

# X is features array, y is outcomes
X = combined_results.drop(columns=['Date','Home Team','Away Team','FT Result','HT Result','Referee',
                                  'FT Home Goals','FT Away Goals','HT Home Goals','HT Away Goals']) # drop non-numerical features
Y = combined_results[['FT Result']]
y = np.ravel(Y,order='C') # flatten the outcome variable
```

```python
# Split the features and outcomes arrays into train and test data
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y)
print(Xtrain.shape, ytrain.shape)
print(Xtest.shape, ytest.shape)
```

```python
# Fit the model on training data
logfit = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial')
logfit.fit(Xtrain,ytrain)
y_model = logfit.predict(Xtest)
```

```python
# Use confusion matrix to evaluate performance
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(ytest, y_model)
cnf_matrix
```

```python
# Use accuracy score to see percent correct predictions

# Import accuracy_score to predict results
from sklearn.metrics import accuracy_score  
accuracy_score(ytest, y_model)
```

## ~53-58% prediction accuracy using static stats

## To see if Home/Away bias is accounted for in model, we can compare the predicted results to the actual results and see if they line up

### Historical rates are as follows- Home: 46% / Away 29% / Draw 25%

```python
total_matches = len(ytest)

# Actual results are in the ytest array. We examine breakdown of H/A/D
actual_home = np.count_nonzero(ytest == 'H') 
actual_away = np.count_nonzero(ytest == 'A')
actual_draw = np.count_nonzero(ytest == 'D')

print("Actual results:","Home:",actual_home/total_matches,"Away:",actual_away/total_matches,"Draw:",actual_draw/total_matches)

# Print new line
print() 

# Model results are in the y_model array. We examine breakdown of H/A/D
model_home = np.count_nonzero(y_model == 'H') 
model_away = np.count_nonzero(y_model == 'A')
model_draw = np.count_nonzero(y_model == 'D')

print("Model results:","Home:",model_home/total_matches,"Away:",model_away/total_matches,"Draw:",model_draw/total_matches)
```

## 2. Dynamic Model: transforming static stats to moving averages, in order to predict results beforehand

```python
# Start with the original combined_results table and derive new features
print(combined_results.head())

print()

print(X.head())
```

```python
# Create a function that selects all results (home and away) by team

def get_results(team_name):
    return combined_results.loc[(combined_results['Home Team'] == team_name) | (combined_results['Away Team'] == team_name)]

get_results('Man United')
```

```python
# Using the get_results formula, generate moving averages

# For each row, use get_results for Home Team, get_results for Away team and create two new columns:
## 1. Home team past 5 games goals
## 2. Away team past 5 games goals
```

```python
def recent_goals(team_name):
    '''This function calculates the average # of goals scored in the last 5 games, based on the input team'''
    
    recent_matches = get_results(team_name)
    
    recent_matches['Last 5 Goals'] = recent_matches['FT Home Goals'].rolling(5).mean()
    
    return recent_matches
    
recent_goals('Man United')
```
